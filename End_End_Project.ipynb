```python
import re
from transformers import AutoTokenizer
# Function to split the text into chunks of a maximum number of tokens.
# Inspired by OpenAI
def overlapping_chunks(text, max_tokens=500, overlapping_factor=5):
  """
  Splits a text into chunks of a maximum number of tokens with overlapping sentences.
  Args:
    text: The text to split.
    max_tokens: The maximum number of tokens per chunk.
    overlapping_factor: The number of sentences to start each chunk with that overlaps wit
  Returns:
    A list of text chunks.
  """
  # Load a tokenizer
  tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
  # Split the text using punctuation
  sentences = re.split(r' [.?!]', text)
  # Get the number of tokens for each sentence
  n_tokens = [len(tokenizer.encode(sentence)) for sentence in sentences]
  chunks, tokens_so_far, chunk = [], [], []
  # Loop through the sentences and tokens joined together in a tuple
  for sentence, tokens in zip(sentences, n_tokens):
    # If the current chunk plus the current sentence exceeds the max tokens, start a new c
    if tokens_so_far + tokens > max_tokens:
      chunks.append(" ".join(chunk))
      chunk = []
      tokens_so_far = 0
    # Add the current sentence to the current chunk
    chunk.append(sentence)
    tokens_so_far += tokens
    # If this is not the first chunk, add the previous overlapping_factor sentences to the
    if len(chunks) > 0:
      for i in range(overlapping_factor):
        if len(chunk) > i:
          chunk[i] = sentences[len(chunks) * overlapping_factor + i]
  # Append the final chunk
  chunks.append(" ".join(chunk))
  return chunks
```
