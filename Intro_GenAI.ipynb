```python
from transformers import Pipeline
```

```python
summarizer = pipeline("summarization")

text = """The giant squid is a large, deep-sea cephalopod mollusk that is found in all oceans of the world. It is one of the largest invertebrates, with specimens reaching up to 18 meters (59 ft) in length. The giant squid has a mantle (body) that is long and cylindrical, and eight long arms that are lined with suckers. It also has two shorter tentacles that are used for grasping prey. The giant squid is a carnivore and feeds on fish, crustaceans, and other marine animals. It is a solitary creature and is rarely seen by humans.

Despite its size, the giant squid is a relatively gentle creature. It is not known to attack humans, and its primary defense mechanism is to squirt ink into the water to cloud the vision of predators. The giant squid is an important part of the marine ecosystem, and its populations are thought to be declining due to overfishing and habitat destruction."""

summary = summarizer(text)

print(summary)
```
